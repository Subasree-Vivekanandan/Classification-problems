## Logistic Regression

predict a categorical dependent variable from a number of independent variables.

Calculate Maximum likelihood : best curve that fits our data

MLE - https://polaris000.medium.com/understanding-maximum-likelihood-estimation-e63dff65e5b1

Likelihood is not a probability density function, meaning that integrating over a specific interval would not result in a “probability” over that interval

Rather, it talks about how likely a distribution with certain values for its parameters fits our data.

## Loading data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
X = df.drop("Purchased",axis=1).values
y = df.Purchased.values
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(2)
X_train,X_test,y_train,y_test = train_test_split(X,
                                                 y,
                                                 test_size=0.3)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
sc_x = StandardScaler()

log_reg = LogisticRegression()
X_scaled = sc_x.fit_transform(X_train)

log_reg.fit(X_scaled,y_train)
```

```python
X_train
```

```python
X_scaled
```

```python
X_test = sc_x.transform(X_test)
```

```python
X_test
```

```python
logreg_ypreds = log_reg.predict(X_test)
logreg_ypreds
```

```python
logreg_ypreds.shape
```

```python
y_test.shape
```

```python
log_reg_Score = log_reg.score(X_test,y_test)
log_reg_Score
```

```python
from sklearn.metrics import accuracy_score


accuracy_score(y_test,logreg_ypreds)
```

```python
log_reg
```

```python
log_reg.predict(sc_x.transform([[30,87000]]))
```

```python
pd.crosstab(y_test,logreg_ypreds,rownames=["Actuallael"],colnames=["Predicted label"])
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test,logreg_ypreds),annot=True,cmap="copper",cbar=False);
plt.xlabel("Predicted label")
plt.ylabel("Actual label");
```

## K neigherest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

Knn = KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)

Knn.fit(X_train,y_train)
```

```python
Knn.score(X_test,y_test)
```

```python
knn_preds = Knn.predict(X_test)
knn_preds
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test,knn_preds,colorbar=False);
```

```python
print(Knn.predict([[30,87000]]))
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_train,y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, Knn.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test,y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, Knn.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
